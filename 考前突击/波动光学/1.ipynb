$$m \ddot{x} = F$$

这是一个二阶常微分方程，我们可以通过两次积分求解：

$$\dot{x} = \frac{F}{m}t + v_0 $$

$$x = \frac{F}{2m}t^2 + v_0 t + x_0$$

为了方便起见，我们定义$ k = \frac{F}{m}$ ，则上面的方程可以写成：

$$\dot{x} = kt + v_0$$

$$x = \frac{k}{2}t^2 + v_0 t + x_0 $$
